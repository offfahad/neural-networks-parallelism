```python
import numpy as np
import matplotlib.pyplot as plt

N = 4
ind = np.arange(N)  # the x locations for the groups
width = 0.25       # the width of the bars

cpu_local_times = (0, 0.05, 0.43, 0.55)
cpu_dist_times = (0.07, 0.03, 0.22, 0.38)

xlabels = ('scatter', 'Forward Propagation', 'Back Propagation', 'Total')

for i in range(1, N):
    fig, ax = plt.subplots()

    ind = 0.0
    scalar_base = min(cpu_local_times[i], cpu_dist_times[i])
    rects1 = ax.bar(ind + width*0, cpu_local_times[i]/scalar_base, width, color='indianred')
    rects2 = ax.bar(ind + width*1, cpu_dist_times[i]/scalar_base,  width, color='royalblue')

    # add some text for labels, title, and axes ticks
    ax.set_xticks([ind + width/2])
    ax.set_xticklabels([xlabels[i]], fontsize=14)
    ax.set_ylabel('Execution Time', fontsize=14)
    ax.set_title('Data parallelism performance comparison', fontsize=16)
    ax.legend(
        (rects1[0], rects2[0]),
        ('Single CPU', 'Parallelism on 2 CPUs'),
        loc='upper right'  # Adjusted to upper-right corner
    )

    def autolabel(rects):
        # attach some text labels
        for rect in rects:
            height = rect.get_height()
            ax.text(
                rect.get_x() + rect.get_width()/2.,
                1.0*height,
                '%.2f' % height,
                ha='center', va='bottom', fontsize=14
            )

    autolabel(rects1)
    autolabel(rects2)

    param = plt.gcf()
    pltsize = param.get_size_inches()
    param.set_size_inches(8, 6)  # Adjusted size, you can change it accordingly
    param.savefig('outputs/' + xlabels[i].lower().replace(' ', '-'))
    plt.show()
```
